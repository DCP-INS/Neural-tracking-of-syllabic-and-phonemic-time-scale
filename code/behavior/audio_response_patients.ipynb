```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.lines as mlines
from scipy import stats
```

```python
# read behavioral data
df = pd.read_csv("data\\behavior\\repeated_wordsfinal.csv")
```

```python
coloris = ["#abbdff","mistyrose","#adc4ad","#6093ff","lightcoral","#5f8c61","#2154a2","#a51b0b","#38713e"]
```

```python
# format data properly
inds = [0, 1, 2, 4, 5, 6, 8,9,10]
conds = [31,61,91, 32,62,92, 33, 63,93]
doto = []
yrrs = []
for num,i in enumerate(conds):
    doto.append(np.mean(df.loc[df['conditions']==i]['%'].values,0))
    yrrs.append(stats.sem(df.loc[df['conditions']==i]['%'].values))
#coloris = ["#a4c2f4","#f4cccc","#d9ead3","#3c78d8", "#e06666","#6aa84f","#1c4587","#990000","#274e13"]
coloris = ['paleturquoise',"mistyrose","yellowgreen","dodgerblue","lightcoral","green","#2154a2","darkred","darkslategray"]
```

```python
# plot data
custom_lines = [mlines.Line2D([0], [0], color="#3c78d8", lw=4),
                mlines.Line2D([0], [0], color="#e06666", lw=4),
                mlines.Line2D([0], [0], color="#6aa84f", lw=4)]

plt.rcParams.update({'font.size': 24})
fig,axs = plt.subplots(1,2,figsize=(16,9),gridspec_kw={'width_ratios': [4, 2]})
ax = axs.ravel()[0]
ax.bar(inds,doto,yerr= yrrs,align='center',color=coloris)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_ylabel('correctly repeated words [%]')
ax.set_xlabel('Phonemic time scale');
ax.set_xticks([1,5,9]);
ax.set_xticklabels(['low','medium','high'],rotation=45);
axs.ravel()[1].legend(custom_lines,['3Hz', '6Hz', '9Hz'],title='Syllabic time scale',fontsize=24)
axs.ravel()[1].set_axis_off()

#fig.savefig( "behavior.png",  dpi=300, metadata=None,bbox_inches='tight',  facecolor=None, edgecolor=None )
```
